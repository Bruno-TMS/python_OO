```python
from decorator import getfullargspec, decorator
from time import sleep, strftime
```

```python
@decorator
def logar(f, fmt='%H:%M:%S', *args, **kwargs):
    print(strftime(fmt))
    return f(*args, **kwargs)
```

```python
@logar
def mochileiro():
    return 42
```

```python
print(mochileiro())
```
