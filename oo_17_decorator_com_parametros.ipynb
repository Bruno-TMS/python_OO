```python
from time import sleep, strftime
from functools import wraps
```

```python
def logar(fmt='%H:%M:%S'):
    def decorator(f):
        @wraps(f)
        def executar_com_tempo(*args, **kwargs):
            print(strftime(fmt))
            f(*args, **kwargs)
    
    return decorator
    
```

```python
@logar(fmt='%d/%M/%Y %H:%M:%S')
def ola(nome):
  return f'Olá {nome}'

@logar(fmt='%d/%M %H:%M:%S')
def mochileiro():
    return 42
```

```python
print(mochileiro())
sleep(3)
print(ola('Renzo'))
```
