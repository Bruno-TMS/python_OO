```python
def fabrica_de_contador(nome):
    _contador = 0
    
    def contar():
        nonlocal _contador
        _contador += 1
        return _contador
    
    contar.nome = nome
    
    return contar
```

```python
contador = fabrica_de_contador('contador 1')
contador_2 = fabrica_de_contador('contador 2')
```

```python
def print_func(f):
    print(f'{f.nome}: {f()}')
```

```python
print_func(contador)
print_func(contador)
print_func(contador)
print_func(contador_2)
print_func(contador_2)
print_func(contador_2)
print_func(contador)
print_func(contador_2)
print_func(contador)
print_func(contador_2)
```
