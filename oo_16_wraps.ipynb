```python
from time import sleep, strftime
from functools import wraps
```

```python
def ola_com_timer_interno(nome):
    print(strftime('%H:%M:%S'))
    return f'Olá {nome}'

def mochileiro_com_timer_interno():
    print(strftime('%H:%M:%S'))
    return 42
```

```python
print(mochileiro_com_timer_interno())
sleep(3)
print(ola_com_timer_interno('Renzo'))
```

```python
def logar(f):
    @wraps(f)
    def executar_com_tempo(*args, **kwargs):
        print(strftime('%H:%M:%S'))
        return f(*args, **kwargs)
    
    return executar_com_tempo
    
```

```python
@logar
def ola(nome):
    print(f'Olá {nome}')

@logar
def mochileiro():
    return 42
```

```python
print(ola.__name__)
print(mochileiro.__name__)
```

```python
print(mochileiro())
sleep(3)
print(ola('Renzo'))
```
